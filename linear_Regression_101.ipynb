```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
# Create features
x = np.linspace(-5,20,10)
# Create Output
y = np.linspace(0,50,10)
# Visualize it
plt.scatter(x, y);
```

```python
mymodel = tf.keras.Sequential([tf.keras.layers.Input(shape = (1,)),
                               tf.keras.layers.Dense(1)])
mymodel.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.SGD(learning_rate=(0.02)),
                metrics="mae")
tf.random.set_seed(40)
```

Creating Models

```python
r=mymodel.fit(x,y,epochs=500)
```

```python
plt.figure(1)
plt.plot(r.history["loss"])
plt.figure(2)
yp=mymodel.predict(x)
plt.scatter(x,y)
plt.plot(x,yp)
```

```python
np.random.seed(101)
# Generating random linear data
# There will be 50 data points ranging from 0 to 50
x = np.linspace(50, 150, 50)
y = np.linspace(0, 50, 50)
# Adding noise to the random linear data
x += np.random.uniform(-4, 4, 50)
y += np.random.uniform(-4, 4, 50)
plt.scatter(x, y)
plt.xlabel('x')
plt.xlabel('y')
plt.title("Training Data")
plt.show()
```

```python
model=tf.keras.Sequential([tf.keras.layers.Input(shape=(1,)),
tf.keras.layers.Dense(1)])
model.compile(loss='mae',optimizer=tf.keras.optimizers.Adam(learning_rate=0.008))
def schedule(epoch, lr):
    if epoch >= 300:
      return 0.002
    return 0.2
scheduler = tf.keras.callbacks.LearningRateScheduler(schedule)
r = model.fit(x, y, epochs=1000, callbacks=[scheduler])

```

```python
plt.plot(r.history['loss'])
plt.figure(2)
plt.scatter(x,y);
# xp=np.linspace(50,150,50)
model.summary()
yp=model.predict(x)
# plt.scatter(x,yp);
plt.plot(x,yp,'r');
model.layers[0].get_weights()
model.get_weights()
```
